## MODELO DE DATOS



```python
import sqlite3
import pandas as pd
```

```python
# Seleccionar base de datos

conn = sqlite3.connect('mi_base_de_datos.sqlite')

cursor = conn.cursor()
```

```python
# Consulta para mostrar por completo la tabla de de parques

df = pd.read_sql_query('SELECT * FROM parque_atracciones', conn)
df
```

```python
# Consulta para mostrar por completo la tabla de de atracciones

df = pd.read_sql_query('SELECT * FROM atracciones', conn)
df
```

```python
# Consulta para mostrar por completo la tabla de Zonas

df = pd.read_sql_query('SELECT * FROM zonas', conn)
df
```

```python
# Consulta para mostrar por completo la tabla de Alojamientos

df = pd.read_sql_query('SELECT * FROM alojamiento', conn)
df
```

```python
# Atracciones aptas para nios y con altura > 120cm

query='''SELECT * FROM atracciones
WHERE Altura_Minima>=120 and Tipo="Infantil"
'''

df = pd.read_sql_query(query, conn)
df
```

```python
# Mostras los alojamientos que hay en cada parque.

query='''SELECT *
FROM alojamiento as A
INNER JOIN Parque_atracciones as B ON A.codigo_parque=B.codigo
'''

df = pd.read_sql_query(query, conn)
df
```

```python
# Mostras los alojamientos de 5 estrellas que hay en cada parque.
query='''SELECT *
FROM alojamiento as A
INNER JOIN Parque_atracciones as B ON A.codigo_parque=B.codigo
Where Categoria = '5 estrellas'
'''

df = pd.read_sql_query(query, conn)
df
```

```python
# Los 5 alojamientos mas baratos en los 3 parques. Mostrando las columnas mar importantes.
query='''SELECT A.Nombre, A.Precio, A.Categoria, B.Nombre
FROM alojamiento as A
INNER JOIN Parque_atracciones as B ON A.codigo_parque=B.codigo
Order by Precio
Limit 5
'''

df = pd.read_sql_query(query, conn)
df
```

```python
# Atracciones para nios mayores de 12 y en que parques las encontramos.

query='''SELECT A.Nombre, A.Tipo, A.Categoria_edad, A.Altura_Minima, B.Nombre
FROM atracciones as A
INNER JOIN Parque_atracciones as B ON A.codigo_parque=B.codigo
Where Categoria_edad>12
'''

df = pd.read_sql_query(query, conn)
df
```
